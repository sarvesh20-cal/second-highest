```python
df=pd.read_excel("/content/EXP1Outliers.xlsx")
print (df)
```

```python
mean_age = df['AGE'].mean()
df['AGE'] = df['AGE'].astype(float)
df.loc[df['AGE'] > 40, 'AGE'] = mean_age
print(df)
```

```python
lower=20
upper=40
mean_value=df["AGE"].mean()
df.loc[(df['AGE']<lower)| (df['AGE']>upper),'MARKS']=mean_value

print (df)
```

```python
lower=20
upper=40
median_value=df["AGE"].median()
df.loc[(df['AGE']<lower)| (df['AGE']>upper),'MARKS']=median_value

print (df)
```

```python
lower=20
upper=40
mode_value=df["AGE"].mode()
df.loc[(df['AGE']<lower)| (df['AGE']>upper),'MARKS']=mode_value
print (df)
```

```python
df['MARKS']=(df['MARKS']-df['MARKS'].min())/(df['MARKS'].max()-df['MARKS'].min())
print(df)

if 'AGE_normalized' in df.columns:
    df = df.drop(columns=['AGE_normalized'])
    print("\nRemoved 'AGE_normalized' column.")
    print(df.head())
```
