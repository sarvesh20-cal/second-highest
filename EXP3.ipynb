```python
import math

def calculate_pass_fail(hours, b0, b1):
    z = b0 + (b1 * hours)

    probability = 1 / (1 + math.exp(-z))

    status = "PASS" if probability >= 0.5 else "FAIL"

    print(f"Hours: {hours}")
    print(f"  Calculation: z = {b0} + ({b1} * {hours}) = {z}")
    print(f"  Probability: P = 1 / (1 + e^-({z})) = {probability:.4f}")
    print(f"  Final Result: {status}\n")

beta_0 = -4
beta_1 = 1

calculate_pass_fail(7, beta_0, beta_1)
calculate_pass_fail(3, beta_0, beta_1)
```
